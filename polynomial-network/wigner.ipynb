```python
import qutip
from qutip import coherent_dm, squeeze, fock_dm, wigner, displace

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

%load_ext autoreload
%autoreload 2
```

# 3D Representation

```python
rho_coherent = coherent_dm(2, 0)
rho_squeeze = squeeze(2,3)
rho_fock = fock_dm(3,1)
xvec = np.linspace(-5,5,200)

W_coherent = wigner(rho_coherent, xvec, xvec)
W_squeeze = wigner(rho_squeeze, xvec, xvec)
W_fock = wigner(rho_fock, xvec, xvec)
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
plt.rcParams['figure.figsize'] = (12,7)

ax1 = fig.add_subplot(111, projection='3d')
ax1.contour3D(xvec, xvec, W_coherent, 300)
plt.xlabel('X')
plt.ylabel('P')
plt.title("W(X,P)")
# plt.savefig("../images/wigner-vacuum.png")
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
plt.rcParams['figure.figsize'] = (12,7)

ax1 = fig.add_subplot(111, projection='3d')
ax1.contour3D(xvec, xvec, W_fock, 300)
plt.xlabel('X')
plt.ylabel('P')
plt.title("W(X,P)")
# plt.savefig("../images/wigner-fock-1.png")
```

# 2D Representation

## Coherent states

```python
plt.rcParams['figure.figsize'] = (8,8)
xvec = np.linspace(-6,6,500)
rho_coherent = coherent_dm(60, 1)
# rho_coherent = coherent_dm(20, 0)

W_coherent = wigner(rho_coherent, xvec, xvec)

plt.axis('equal')
plt.axhline(linewidth=1, zorder=2, color="white")
plt.axvline(linewidth=1, zorder=2, color="white")
plt.contourf(xvec, xvec, W_coherent, 300, zorder=1)
plt.xlabel("X")
plt.ylabel("P")
plt.axis((-5,5,-5,5))
plt.tight_layout()
plt.margins(0,0)
plt.grid()
# plt.grid()
# plt.savefig("../images/wigner-coherent-1.png")
```

```python
N = 40

plt.rcParams['figure.figsize'] = (8,8)
xvec = np.linspace(-6,6,500)
rho_squeezing = coherent_dm(N, 0)
rho_squeezing = squeeze(N,0.5 + 0.5j)(rho_squeezing)
# rho_squeezing = squeeze(N,0.5 + 0.5j) * rho_squeezing * squeeze(N,0.5 + 0.5j).dag()
# rho_squeezing = displace(N, 1.5 + 1.5j)(rho_squeezing)
# rho_coherent = coherent_dm(20, 0)

W_squeezing = wigner(rho_squeezing, xvec, xvec)

plt.axis('equal')
plt.axhline(linewidth=1, zorder=2, color="white")
plt.axvline(linewidth=1, zorder=2, color="white")
plt.contourf(xvec, xvec, W_squeezing, 300, zorder=1)
plt.xlabel("X")
plt.ylabel("P")
plt.axis((-5,5,-5,5))
plt.tight_layout()
plt.margins(0,0)
# plt.grid()
# plt.savefig("../images/wigner-squeezed-2.png")
```

## X and P eigenstates

```python
N = 70

plt.rcParams['figure.figsize'] = (8,8)
xvec = np.linspace(-6,6,500)
rho_squeezing = coherent_dm(N, 0)
rho_squeezing = squeeze(N,-2)(rho_squeezing)
rho_squeezing = displace(N, 1j)(rho_squeezing)
# rho_coherent = coherent_dm(20, 0)

W_squeezing = wigner(rho_squeezing, xvec, xvec)
W_squeezing[W_squeezing <= 0.1] = 0
# W_squeezing[W_squeezing > 0.1] = 1

plt.axis('equal')
plt.axhline(linewidth=1, zorder=2, color="white")
plt.axvline(linewidth=1, zorder=2, color="white")
plt.contourf(xvec, xvec, W_squeezing, 300, zorder=1)
plt.xlabel("X")
plt.ylabel("P")
plt.axis((-5,5,-5,5))
plt.tight_layout()
plt.margins(0,0)
# plt.grid()
plt.savefig("../images/wigner-p-eigenstate.png")
```
